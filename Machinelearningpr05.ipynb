###  K- Nearest Neibours


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

### allocate x and y dependent and inependent variables

```python
 x=df.iloc[:,0:8]
 y=df.iloc[:,-1]
```

```python
y[0:5]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(x).transform(x)
```

```python
df.shape
```

### Train test spilt

```python
from sklearn.model_selection import  GridSearchCV,train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

### importing the library and creating model for KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = 7
#Train Model and Predict  
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
neigh
```

### predicting the values

```python
yhat = neigh.predict(X_test)
yhat[0:5]
```

```python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat))
```

```python

k = 11
neigh6 = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
yhat6 = neigh6.predict(X_test)
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh6.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat6))

```

```python
Ks = 20
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
```

```python
print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1) 
```

### creating kfolds and gridsearchcv

```python
knn_classifier = KNeighborsClassifier()
```

```python
param_grid = {'n_neighbors': np.arange(1, 21),  
              'weights': ['uniform', 'distance']}
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
grid_search = GridSearchCV(knn_classifier, param_grid, cv=kfold, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_score = grid_search.best_score_
```

```python
y_pred = grid_search.best_estimator_.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')  
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
```

```python
print("Best Parameters:", best_params)
print("Best Accuracy:", best_score)
print("Test Set Metrics:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
