### Splitting data into training and testing set

```python
msk=np.random.rand(len(df))>0.8
train=ftr[msk]
test=ftr[~msk]
```

### Training the multiple linear regression model using the trainnig set

```python
from sklearn import linear_model
regression=linear_model.LinearRegression()
train_x=np.asanyarray(train[['R&D Spend','Administration','Marketing Spend']])
train_y=np.asanyarray(train[['Profit']])
regression.fit(train_x,train_y)
```

```python
print('coefficient',regression.coef_)
print('intercept',regression.intercept_)
```

### Testing the model and finding the errors in the model

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['R&D Spend','Administration','Marketing Spend']])
test_y = np.asanyarray(test[['Profit']])
test_y_ = regression.predict(test_x)
```

```python
from sklearn.metrics import mean_absolute_error 
mean_absolute_error(test_y,test_y_)
```

```python
from sklearn.metrics import mean_squared_error 
mean_squared_error(test_y,test_y_)
```

```python
from sklearn.metrics import root_mean_squared_error 
root_mean_squared_error(test_y,test_y_)
```

```python
print("R2-score: %.2f" % r2_score(test_y , test_y_) )
```
